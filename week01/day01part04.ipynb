```python
#Let's learn Lists!!!
# LISTS
# Ordered, mutable collection of elements.
# Elements are stored in square brackets [].

numbers = [10, 20, 30, 40, 50]

# Access elements using index
# Index starts at 0
print(numbers[0])   # 10
print(numbers[2])   # 30

# Negative indexing (from the end)
print(numbers[-1])  # 50
print(numbers[-2])  # 40
```

```python
#List Slicing
# LIST SLICING
# Syntax: list[start:end]
# Start index included, end index NOT included.

nums = [10, 20, 30, 40, 50, 60]

print(nums[1:4])   # [20,30,40]

# If start is omitted, Python assumes start = 0
print(nums[:3])    # [10,20,30]

# If end is omitted, Python slices until the end
print(nums[2:])    # [30,40,50,60]

# Negative slicing
print(nums[-3:])   # [40,50,60]

# Step parameter
# Syntax: list[start:end:step]

print(nums[::2])   # [10,30,50]

# Reverse a list
print(nums[::-1])  # [60,50,40,30,20,10]
```

```python
#List methods
numbers = [3, 1, 5]

# append() adds element at end
numbers.append(7)

# insert(index, value)
numbers.insert(1, 2)

# remove(value)
numbers.remove(5)

# pop() removes last element
numbers.pop()

# sort()
numbers.sort()

# reverse()
numbers.reverse()

# length
len(numbers)
```

```python
#Tuples
# TUPLES
# Ordered but immutable collection.
# Cannot modify elements after creation.

coords = (10, 20)

print(coords[0])   # 10
print(coords[1])   # 20

# Tuple slicing works like lists
t = (1,2,3,4)
print(t[1:3])  # (2,3)

# Tuple unpacking
a, b = (100, 200)

print(a)  # 100
print(b)  # 200

# Tuples cannot be modified
# t[0] = 5 → TypeError
```

```python
# DICTIONARIES
# Store key-value pairs.
# Created using {}

person = {
    "name": "CJ",
    "age": 30,
    "city": "Waterloo"
}

# Access values using keys
print(person["name"])   # CJ

# get() safely retrieves value
print(person.get("age"))  # 30

# Adding new key
person["email"] = "cj@email.com"

# Updating value
person["age"] = 31

# Remove key
person.pop("city")

# Retrieve keys
person.keys()

# Retrieve values
person.values()

# Retrieve key-value pairs
person.items()
```

```python
# SETS
# Unordered collection of unique elements.
# Duplicate values automatically removed.

numbers = {1,2,3,3,4}

print(numbers)   # {1,2,3,4}

# Add element
numbers.add(5)

# Remove element
numbers.remove(3)
```

```python
a = {1,2,3}
b = {3,4,5}

# Union (combine unique elements)
print(a | b)   # {1,2,3,4,5}

# Intersection (common elements)
print(a & b)   # {3}

# Difference
print(a - b)   # {1,2}
```

```python
# 1. CONDITIONAL STATEMENTS
# Conditional statements allow programs to make decisions.

x = 10

if x > 10:
    print("Greater than 10")

elif x == 10:
    print("Equal to 10")

else:
    print("Less than 10")


# Comparison operators
# ==  equal
# !=  not equal
# >   greater than
# <   less than
# >=  greater than or equal
# <=  less than or equal


# Logical operators
# and  -> both conditions must be True
# or   -> at least one condition must be True
# not  -> reverses a boolean value

x = 15

if x > 10 and x < 20:
    print("x is between 10 and 20")
```

```python
# ------------------------------------------
# 3. NESTED IF STATEMENTS
# ------------------------------------------
# A nested if means an if statement inside another if.

age = 20

if age >= 18:
    if age >= 21:
        print("Can drink in US")
    else:
        print("Adult but cannot drink in US")
else:
    print("Minor")


# Nested if is useful when a second condition
# should only be checked if the first condition is True.
```

```python
# ------------------------------------------
# 4. SHORT-CIRCUIT EVALUATION
# ------------------------------------------
# Python sometimes stops evaluating logical conditions early.

# AND short circuit
# If the first condition is False, Python does not check the second.

x = -5

if x > 0 and x > 10:
    print("Positive number")


# OR short circuit
# If the first condition is True, Python does not check the second.

x = 15

if x > 10 or x > 100:
    print("Condition satisfied")


# Example preventing errors

x = None

if x is not None and x > 5:
    print("Valid number")
```

```python
# ------------------------------------------
# 7. ENUMERATE
# ------------------------------------------
# enumerate() returns index + value while looping.

names = ["Alice", "Bob", "Charlie"]

for index, name in enumerate(names):
    print(index, name)


# enumerate produces pairs like:
# (0, "Alice")
# (1, "Bob")
# (2, "Charlie")
```

```python
# ------------------------------------------
# 8. ZIP
# ------------------------------------------
# zip() combines elements from multiple lists.

names = ["Alice", "Bob"]
ages = [25, 30]

for name, age in zip(names, ages):
    print(name, age)


# zip produces pairs like:
# ("Alice", 25)
# ("Bob", 30)
```

```python
import random

# Random number generated by computer
system_guessed_number = random.randint(1, 100)

max_attempts = 7
guessed_correctly = False

for attempt in range(max_attempts):

    user_input = int(input("Guess a number between 1 and 100: "))

    if user_input > system_guessed_number:
        print("Too high!")

    elif user_input < system_guessed_number:
        print("Too low!")

    else:
        print("You got it!")
        guessed_correctly = True
        score = max_attempts - attempt
        print(f"Score: {score}")
        break

# if user never guessed correctly.
if not guessed_correctly:
    print(f"Better luck next time. The correct number was {system_guessed_number}.")
```
