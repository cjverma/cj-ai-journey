## 1. Nested Dictionaries

```python
# Contact book — nested dictionary
phone_book = {
    'person_a': {'name': 'Chiranjay Verma',  'phone': '+1-519-574-4155', 'email': 'chiranjay.verma@gmail.com'},
    'person_b': {'name': 'Chiranjay1 Verma', 'phone': '+1-519-574-4156', 'email': 'chiranjay.verma1@gmail.com'},
    'person_c': {'name': 'Chiranjay2 Verma', 'phone': '+1-519-574-4157', 'email': 'chiranjay.verma2@gmail.com'}
}
```

```python
# .keys(), .values(), .items() — know what each returns
print(phone_book.keys())    # just the outer keys
print(phone_book.values())  # just the nested dicts
print(phone_book.items())   # both together as pairs
```

```python
# Loop through all contacts and print everything
for person, details in phone_book.items():
    for key, value in details.items():
        print(key, value)
```

```python
# Print ONLY the name of each person — loop + condition
for person, details in phone_book.items():
    for key, value in details.items():
        if key == 'name':
            print(value)
```

```python
# Access one specific value directly — no loop needed
print(phone_book['person_b']['email'])
```

## 2. Functions

```python
# print vs return — understand the difference

# Version 1 — print: shows value but throws it away
def get_email_print(person):
    print(phone_book[person]['email'])

result = get_email_print('person_a')
print('Result is:', result)  # None — nothing was returned
```

```python
# Version 2 — return: hands the value back so you can use it
def get_email_return(person):
    return phone_book[person]['email']

result = get_email_return('person_a')
print('Result is:', result)  # actual email — you can keep using it
```

```python
# Rule: almost always use return. Use print only to display to a user.

# Function with parameters — works for any person, any field
def get_contact_info(person, field):
    return phone_book[person][field]

print(get_contact_info('person_b', 'phone'))
print(get_contact_info('person_c', 'email'))
```

## 3. Sets

```python
# Sets: no duplicates, no order
cities = {'Toronto', 'Mumbai', 'Toronto', 'London', 'Mumbai'}
print(cities)  # duplicates removed automatically
```

```python
# Set operations
morning_class = {'Alice', 'Bob', 'Charlie', 'Diana'}
evening_class = {'Charlie', 'Diana', 'Eve', 'Frank'}

print(morning_class & evening_class)  # intersection — in both
print(morning_class | evening_class)  # union — all, no duplicates
print(morning_class - evening_class)  # difference — in morning but not evening
```

```python
# ML use case: check which features appear in training but not in test data
train_features = {'age', 'salary', 'experience', 'location'}
test_features  = {'age', 'salary', 'experience'}

missing_in_test = train_features - test_features
print('Features missing in test:', missing_in_test)  # location
```

## 4. While Loops

```python
# While loop — runs until condition becomes False
x = 0
while x <= 6:
    print(x)   # print first
    x = x + 1  # then increment — order matters!
```

```python
# ML context: train until loss is low enough
# You don't always know how many epochs you need
loss = 1.0
epoch = 0

while loss > 0.1:
    loss = loss * 0.7   # simulating loss decreasing each epoch
    epoch += 1
    print(f'Epoch {epoch}: loss = {loss:.4f}')

print(f'Training done in {epoch} epochs')
```

## 5. enumerate() and zip()

```python
# enumerate() — loop with index, no manual counter needed
cities = ['Toronto', 'Mumbai', 'London']

# Manual version
x = 0
for city in cities:
    print(x, city)
    x += 1
```

```python
# enumerate() version — cleaner, same result
for index, city in enumerate(cities):
    print(index, city)
```

```python
# zip() — loop over two lists side by side
predictions = [0.9, 0.3, 0.8, 0.1]
actual =      [1,   0,   1,   0  ]

for pred, act in zip(predictions, actual):
    print(pred, act)
```

```python
# ML context: basic model evaluation loop
for pred, act in zip(predictions, actual):
    if pred >= 0.5 and act == 1:
        print('Correct positive')
    elif pred < 0.5 and act == 0:
        print('Correct negative')
    else:
        print('Wrong')
```

## 6. Putting It All Together — Model Evaluation Function

```python
# Function that counts correct predictions
# Uses: def, parameters, zip(), loop, conditionals, return

def count_correct(predictions, actual):
    x = 0                                      # counter starts at 0
    for pred, act in zip(predictions, actual): # pair them up
        if pred >= 0.5 and act == 1:           # correct positive
            x += 1
        elif pred < 0.5 and act == 0:          # correct negative
            x += 1
        else:
            print('Wrong prediction')
    return x                                   # return AFTER the loop

predictions = [0.9, 0.3, 0.8, 0.1]
actual =      [1,   0,   1,   0  ]

correct = count_correct(predictions, actual)
print(f'{correct} out of {len(actual)} correct')
```

## Key Rules to Remember

| Concept | Rule |
|---|---|
| `/` vs `//` | `/` always returns float. `//` returns int (floor). |
| `print` vs `return` | `print` shows and discards. `return` hands back for reuse. |
| `return` placement | Always AFTER the loop, not inside it. |
| Sets | No duplicates, no order. Use for membership and comparisons. |
| `while` vs `for` | `for` = known iterations. `while` = run until condition false. |
| `enumerate()` | Use when you need index + value together. |
| `zip()` | Use when you need to pair two lists side by side. |
