## 1. List Comprehensions
A list comprehension builds a list in one line instead of a for loop.

Formula: `[expression for item in iterable if condition]`

Rule: if you can't read it in one glance, use a for loop instead. Comprehensions are for clarity, not cleverness.

```python
# The long way
squares = []
for x in range(1, 11):
    squares.append(x**2)
print(squares)
```

```python
# List comprehension — same result, one line
squares = [x**2 for x in range(1, 11)]
print(squares)
```

```python
# With filter — only even squares
even_squares = [x**2 for x in range(1, 11) if x % 2 == 0]
print(even_squares)
```

```python
# Filter words longer than 4 characters
# expression = x (the word itself), not words (the whole list)
words = ['cat', 'elephant', 'dog', 'python', 'rat', 'tiger', 'ox', 'crocodile']
long_words = [x for x in words if len(x) >= 4]
print(long_words)
```

```python
# Convert Celsius to Fahrenheit
temps = [0, 20, 37, 100]
temps_in_fahrenheit = [x * 9/5 + 32 for x in temps]
print(temps_in_fahrenheit)
```

## 2. Lambda
An anonymous one-line function. Use when the function is simple and only needed once.

Lambda is to functions what list comprehensions are to loops — a shorter form for simple cases.

```python
# Regular function
def double(x):
    return x * 2

# Lambda — identical behaviour
double = lambda x: x * 2

print(double(5))  # 10
```

```python
# Lambda that squares a number
power_number = lambda x: x ** 2
print(power_number(5))  # 25
```

## 3. map()
`map(func, list)` applies a function to every item in a list.

Returns a lazy iterator — wrap in `list()` to see the actual values.

```python
# map() — square every number
numbers = [1, 2, 3, 4, 5]
result = list(map(lambda x: x**2, numbers))
print(result)  # [1, 4, 9, 16, 25]

# Without list() — returns object location, not values
print(map(lambda x: x**2, numbers))  # <map object at 0x...>
```

```python
# map() — convert names to uppercase
names = ['john', 'alice', 'bob', 'diana', 'tom']
uppercase_names = list(map(lambda x: x.upper(), names))
print(uppercase_names)
```

---
## 4. filter()
`filter(func, list)` keeps only items where the function returns True.

map() transforms every item. filter() keeps only items that pass a condition.

```python
# filter() — keep only numbers greater than 3
numbers = [1, 2, 3, 4, 5]
result = list(filter(lambda x: x > 3, numbers))
print(result)  # [4, 5]
```

```python
# filter() — keep only names longer than 4 characters
names = ['john', 'alice', 'bob', 'diana', 'tom']
long_names = list(filter(lambda x: len(x) > 4, names))
print(long_names)
```

---
## 5. Chaining map() and filter()

```python
# Chain: filter first, then map — all in one expression
# filter() output is passed directly as input to map()
names = ['john', 'alice', 'bob', 'diana', 'tom']
result = list(map(lambda x: x.upper(), filter(lambda x: len(x) > 4, names)))
print(result)  # ['ALICE', 'DIANA']
```

---
## Key Rules to Remember

| Concept | Rule |
|---|---|
| List comprehension | `[expression for item in iterable if condition]`. Expression = what each item becomes. |
| `x` vs `words` | In a comprehension, `x` is one item. `words` is the whole list. Always use `x`. |
| Lambda | One-line anonymous function. Use for simple, single-use operations. |
| `map()` | Transforms every item. Returns lazy iterator — wrap in `list()`. |
| `filter()` | Keeps items where function returns True. Returns lazy iterator — wrap in `list()`. |
| Chaining | Pass `filter()` directly as second argument to `map()` — one expression, no intermediate variable. |
