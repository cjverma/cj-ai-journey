## Core Concepts

- `def` declares a function
- `return` sends a value back — function exits immediately when return is hit
- A function with no return gives back `None`
- `return` placement matters — inside a loop exits early, after a loop returns the final result
- Variables defined inside a function stay inside — accessing them outside causes `NameError`
- A comparison like `n % 2 == 0` already returns `True` or `False` — no need to wrap in a ternary

---
## Function 1 — is_even()

````python
# Returns True if n is even, False if odd
# Key insight: n % 2 == 0 already IS a boolean — return it directly

def is_even(n):
    return n % 2 == 0

print(is_even(4))   # True
print(is_even(25))  # False
````

---
## Function 2 — celsius_to_fahrenheit()

````python
# Formula: (c × 9/5) + 32
# Test with -40 — Celsius and Fahrenheit meet at exactly -40

def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

print(celsius_to_fahrenheit(-40))  # -40.0
print(celsius_to_fahrenheit(100))  # 212.0
print(celsius_to_fahrenheit(0))    # 32.0
````

---
## Function 3 — count_vowels()

````python
# Returns how many vowels are in a string
# Key lessons from building this:
# 1. x = 0 must be INSIDE the function — outside causes UnboundLocalError
# 2. return must be AFTER the loop — inside the loop exits after first vowel found
# 3. char in 'aeiou' uses the same 'in' operator as list membership testing

vowels = 'aeiou'

def count_vowels(b):
    x = 0                # counter inside function — resets each call
    for i in b:
        if i in vowels:  # 'in' works on strings too
            x = x + 1
    return x             # after loop — returns final count

print(count_vowels('hello'))       # 2
print(count_vowels('aeiouaeiou'))  # 10
print(count_vowels('rhythm'))      # 0
````

---
## Key Rules to Remember

| Concept | Rule |
|---|---|
| `return` vs `print` | `return` sends value back for reuse. `print` shows and discards. |
| `return` placement | Always AFTER the loop. Inside the loop = exits after first iteration. |
| Variable scope | Variables inside a function stay inside. Define counter inside, not outside. |
| Boolean shortcut | `n % 2 == 0` is already True/False — return it directly, no ternary needed. |
| `in` operator | Works on both lists AND strings — `char in 'aeiou'` is valid. |
