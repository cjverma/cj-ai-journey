## 1. range() — Three Variations

```python
# range(stop) — starts at 0
for i in range(5):
    print(i)  # 0 1 2 3 4
```

```python
# range(start, stop) — starts at start, stops before stop
for i in range(1, 6):
    print(i)  # 1 2 3 4 5
```

```python
# range(start, stop, step) — jumps by step each time
for i in range(0, 10, 2):
    print(i)  # 0 2 4 6 8
```

## 2. Looping Over Lists and Strings

```python
# Loop over a list
cities = ['Toronto', 'Mumbai', 'London']
for city in cities:
    print(city)
```

```python
# Loop over a string — each character is an item
string = 'hello'
for item in string:
    print(item)  # no index needed — item IS the character
```

## 3. FizzBuzz — Key Insight: Order Matters in if/elif

```python
# FizzBuzz — loop 1 to 30
# Rule: FizzBuzz MUST be checked first
# Why: if divisible by both 3 and 5, it gets caught by FizzBuzz
# and never reaches Fizz or Buzz — no extra conditions needed

for i in range(1, 31):
    if i % 3 == 0 and i % 5 == 0:  # check BOTH first
        print('FizzBuzz')
    elif i % 3 == 0:                # safe — FizzBuzz already handled
        print('Fizz')
    elif i % 5 == 0:                # safe — FizzBuzz already handled
        print('Buzz')
    else:
        print(i)
```

## 4. While Loops

```python
# While loop — runs until condition becomes False
# Always make sure something inside eventually makes condition False

x = 0
while x < 5:
    print(x)   # print first
    x = x + 1  # then increment — order matters!
```

```python
# while/else — else block runs only if loop finishes WITHOUT hitting break
# If break triggers — else is skipped entirely

x = 0
while x < 3:
    print(x)
    x += 1
else:
    print('Loop finished without break')
```

---
## 5. break and continue

```python
# break — exits the loop immediately
for i in range(10):
    if i == 5:
        break       # stops the loop entirely
    print(i)        # prints 0 1 2 3 4
```

```python
# continue — skips rest of current iteration, jumps back to top of loop
for i in range(5):
    if i == 3:
        continue    # skips 3, keeps going
    print(i)        # prints 0 1 2 4
```

## 6. Password Program — Putting It All Together

```python
# Run this in terminal as password_program.py — input() unreliable in Jupyter

# What this covers: while loop, break, continue, while/else, conditionals

correct_password = 'ABCDEFGH'
x = 0

while x < 3:
    input_password = input('What is your password? ')
    x = x + 1
    if input_password == correct_password:
        print('Access granted')
        break
    elif input_password != correct_password and x < 3:  # not last attempt
        print('Try again')
else:
    print('Locked out')  # only runs if break was never triggered
```

## Key Rules to Remember

| Concept | Rule |
|---|---|
| `range()` | Stop is always excluded. Step controls jump size. |
| Loop over string | `for item in string` — item IS the character, no index needed. |
| FizzBuzz | Most specific condition first. Order is your logic. |
| `break` | Exits loop immediately. `while/else` is skipped if break triggers. |
| `continue` | Skips current iteration, jumps back to top of loop. |
| `while/else` | `else` runs only when loop finishes without hitting `break`. |
| Infinite loop | Always make sure something inside the loop eventually makes condition False. |
