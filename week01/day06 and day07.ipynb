## 1. Authentication System — Version 1 (Lists)
First attempt — using lists. Issue: loops through every password, prints wrong message for each non-match.

```python
# Version 1 — lists (flawed but instructive)
password = ["Orange", "Green", "White", "Red", "Blue"]
username = ["cj", "chiranjay", "cverma"]

print("What is your username?")
entered_username = input(">")
for x in username:
    if entered_username == x:
        message = f"What is your password, {x}"
        print(message)
    elif entered_username == x:
        print("Your username is incorrect! System doesn't recognise you")
        break
    entered_password = input(">")
    for y in password:
        if entered_password == y:
            print("Login is successful!")
            break
        elif entered_password != y:
            print("Unfortunately, your password is incorrect!")
            break
```

## 2. Authentication System — Version 2 (Dictionary)
Better approach — username as key, password as value. Direct lookup, no loop needed.

```python
# Version 2 — dictionary lookup + 3 attempt lockout
# Key insight: users[entered_username] gets the password directly — no loop needed

users = {
    "cj": "Orange123",
    "cj1": "Orange1234",
    "cj2": "Orange1235"
}

print("What is your username?")
key = entered_username = input(">")
x = 0

if entered_username in users.keys():
    while x < 3:
        print(f"What is your password, {key}")
        entered_password = input(">")
        if entered_password == users[entered_username]:
            print("Your login is successful.")
            break
        elif x == 2:
            print("Your account is now locked! Please contact customer support!")
            break
        elif entered_password != users[entered_username]:
            print("Your password is incorrect. Try again!")
            x = x + 1
else:
    print("Incorrect Username")
```

## 3. Prime Number Checker
Key lessons: `return` exits function immediately. `for/else` — else runs only if loop finished without `break`.

```python
def prime_number(x):
    if x <= 1:                  # 0 and 1 are never prime — exit immediately
        print("Not a prime number!")
        return
    for n in range(2, x - 1):  # check every number from 2 to x-1
        if x % n == 0:          # divisible — not prime
            print("Not a prime number!")
            break
    else:                       # only runs if loop finished without break
        print("Prime number!")

prime_number(1)   # Not a prime number
prime_number(2)   # Prime number
prime_number(7)   # Prime number
prime_number(9)   # Not a prime number
```

## 4. List Comprehension — Capital Letters

```python
# Return only words that start with a capital letter
# x[0].isupper() checks if first character is uppercase

list_a = ["Orange", "Red", "blue", "green"]
result = [x for x in list_a if x[0].isupper()]
print(result)  # ['Orange', 'Red']
```

## 6. Number Guessing Game — Final Version
First complete standalone program. Run as `guessing_game.py` in terminal — `input()` unreliable in Jupyter.

```python
# Run this in terminal: python guessing_game.py

import random

def check_guess(secret, guess):
    """Returns 'correct', 'high', or 'low' based on guess vs secret."""
    if secret == guess:
        return 'correct'
    elif guess > secret:
        return 'high'
    else:
        return 'low'

def run_game():
    """Main game loop — 6 attempts, input validation, win/lose messages."""
    secret = random.randint(1, 50)
    x = 0
    while x < 6:
        try:
            guess = int(input("What is your guess? "))
        except ValueError:
            print("Please enter a number!")
            continue
        result = check_guess(secret, guess)
        print(result)
        if result == 'correct':
            print(f'It took you {x + 1} attempt(s) to get it right!')
            break
        x = x + 1
    if x == 6:
        print(f'You ran out of attempts! Correct number was {secret}! Try again!')

run_game()
```

## Key Rules from This Week

| Concept | Rule |
|---|---|
| Dictionary vs list for lookup | Use dict when you need to find something by a key — no loop needed. |
| `return` in functions | Exits immediately. Always place AFTER loops, not inside them. |
| `for/else` | `else` runs only if loop finished without hitting `break`. |
| `try/except` | Wraps risky code. `except ValueError` catches bad input like letters instead of numbers. |
| `continue` | Skips rest of iteration, jumps back to top of loop without wasting an attempt. |
| Two functions | `check_guess()` does one thing — returns result. `run_game()` handles all game logic. |
