## Key Rules Locked In Today

| Concept | Rule |
|---|---|
| `/` vs `//` | `/` ALWAYS returns float. `//` returns int (floor). Even `10/5` = `5.0` not `5`. |
| `=` vs `==` | `=` assigns. `==` compares. Never mix them. |
| `sorted()` vs `.sort()` | `.sort()` modifies in place. `sorted()` returns new list, original unchanged. |
| Dictionary `:` vs `=` | Inside a dictionary, key-value pairs use `:`. Variable assignment uses `=`. |
| `in` operator | Works on lists, strings, AND dictionary keys. `'person_a' in people` is valid. |
| `.get()` | Safe dictionary access. `dict.get('key', 'default')` — returns default instead of KeyError. |
| `del` vs `.pop()` | `del people['key']` — removes, no return. `.pop('key')` — removes AND returns value. |

## 1. Lists Drill

```python
# Get last item without hardcoding index
cities = ['Mumbai', 'Toronto', 'Waterloo', 'Delhi', 'New York']
print(cities[-1])  # -1 always gives last item regardless of list length
```

```python
# append() and insert()
cities.append('London')      # adds to end
cities.insert(2, 'Paris')    # inserts at index 2
print(cities)
```

```python
# remove() and pop()
cities.remove('Delhi')       # removes by value
removed = cities.pop()       # removes last item AND returns it
print(removed)               # 'London'
print(cities)
```

```python
# sort() and reverse()
cities.sort()                # sorts in place — modifies original
print(cities)
cities.reverse()             # reverses in place
print(cities)
```

```python
# sorted() — returns NEW list, original unchanged
cities = ['Mumbai', 'Toronto', 'Paris']
new_list = sorted(cities)
print(cities)    # unchanged
print(new_list)  # sorted copy
```

```python
# List comprehensions — filter and transform
cities = ['Mumbai', 'Toronto', 'Paris', 'Waterloo', 'New York']

# Filter only
long_cities = [x for x in cities if len(x) > 6]

# Transform only
upper_cities = [x.upper() for x in cities]

# Filter AND transform — one line
long_upper = [x.upper() for x in cities if len(x) > 6]

print(long_cities)
print(upper_cities)
print(long_upper)
```

```python
# Function with comprehension inside
def get_long_cities(city_list, min_length):
    return [x for x in city_list if len(x) > min_length]

print(get_long_cities(['Mumbai', 'Toronto', 'Paris'], 6))
```

## 2. Dictionaries Drill

```python
# Nested dictionary — correct syntax uses : not =
people = {
    'person_a': {'name': 'CJ1', 'age': 30, 'city': 'Toronto'},
    'person_b': {'name': 'CJ2', 'age': 31, 'city': 'Vancouver'},
    'person_c': {'name': 'CJ3', 'age': 32, 'city': 'Waterloo'}
}
```

```python
# Direct access — no loop needed for single value
print(people['person_b']['city'])  # 'Vancouver'
```

```python
# Loop + condition — one outer loop only, value IS the inner dict
for key, value in people.items():
    if value['age'] > 30:
        print(value['name'])  # no second loop needed
```

```python
# Adding new entry — direct assignment
people['person_d'] = {'name': 'CJ4', 'age': 28, 'city': 'Montreal'}
print(people['person_d'])
```

```python
# Safe access with .get()
# Outer key — safe access
print(people.get('person_e', 'Person not found!'))

# Inner key — safe access when outer key is known to exist
print(people['person_a'].get('phone', 'No phone on record'))
```

```python
# Check if key exists
if 'person_a' in people:
    print(people['person_a'].get('name', "Name doesn't exist"))
```

```python
# Delete a key — two ways
del people['person_d']                    # removes, no return
removed = people.pop('person_c')          # removes AND returns value
print('Removed:', removed)
```

```python
# Three looping methods
people = {
    'person_a': {'name': 'CJ1', 'age': 30, 'city': 'Toronto'},
    'person_b': {'name': 'CJ2', 'age': 31, 'city': 'Vancouver'},
}

for key in people.keys():          # just keys
    print(key)

for value in people.values():      # just values
    print(value)

for key, value in people.items():  # both together
    print(key, value)
```

```python
# Word frequency counter — clean version
text = "the cat sat on the mat the cat"
new_text = text.split()
word_count = {}

for word in new_text:
    if word in word_count:      # key exists — increment
        word_count[word] += 1
    else:                       # key doesn't exist — set to 1
        word_count[word] = 1

print(word_count)

# Sort by frequency descending
sorted_count = dict(sorted(word_count.items(), key=lambda x: x[1], reverse=True))
print(sorted_count)
```
